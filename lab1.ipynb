```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# Base URL of the website
URL = [
    "https://clevershop.mk/product-category/mobilni-laptopi-i-tableti/page/1/",
    "https://clevershop.mk/product-category/mobilni-laptopi-i-tableti/page/2/",
    "https://clevershop.mk/product-category/mobilni-laptopi-i-tableti/page/3/",
    "https://clevershop.mk/product-category/mobilni-laptopi-i-tableti/page/4/",
    "https://clevershop.mk/product-category/mobilni-laptopi-i-tableti/page/5/",
    "https://clevershop.mk/product-category/mobilni-laptopi-i-tableti/page/6/",
    "https://clevershop.mk/product-category/mobilni-laptopi-i-tableti/page/7/",
    "https://clevershop.mk/product-category/mobilni-laptopi-i-tableti/page/8/",
    "https://clevershop.mk/product-category/mobilni-laptopi-i-tableti/page/9/",
    "https://clevershop.mk/product-category/mobilni-laptopi-i-tableti/page/10/",
    "https://clevershop.mk/product-category/mobilni-laptopi-i-tableti/page/11/",
    "https://clevershop.mk/product-category/mobilni-laptopi-i-tableti/page/12/",
    "https://clevershop.mk/product-category/mobilni-laptopi-i-tableti/page/13/",
    "https://clevershop.mk/product-category/mobilni-laptopi-i-tableti/page/14/"
]
```

```python
for i in URL:
  page = requests.get(i)
  print(f'{i} : {page}')
```

```python
data = {
    "Product Title": [],
    "Regular Price": [],
    "Discount Price": [],
    "Product Page URL": [],
    "Add to Cart URL": []
}
```

```python
# Iterate over each page URL
for url in URL:
    page = requests.get(url)
    if page.status_code == 200:
        soup = BeautifulSoup(page.content, 'html.parser')

        # Find all product blocks on the page
        products = soup.find_all('div', class_='product-grid-item')

        for product in products:
            # Product title
            title = product.find('h3', class_='wd-entities-title')
            data["Product Title"].append(title.text.strip() if title else None)

            # Regular and Discount price (if available)
            prices = product.find_all('span', class_='woocommerce-Price-amount')
            if len(prices) > 1:  # Discounted product
                data["Regular Price"].append(prices[0].text.strip())
                data["Discount Price"].append(prices[1].text.strip())
            elif len(prices) == 1:  # Regular priced product
                data["Regular Price"].append(prices[0].text.strip())
                data["Discount Price"].append(None)
            else:  # Price not found
                data["Regular Price"].append(None)
                data["Discount Price"].append(None)

            # Product page URL
            product_page = product.find('a', class_='woocommerce-LoopProduct-link')
            data["Product Page URL"].append(product_page['href'] if product_page else None)

            # Add to Cart URL
            add_to_cart_button = product.find('a', class_='add_to_cart_button')
            data["Add to Cart URL"].append(add_to_cart_button['href'] if add_to_cart_button else None)







```

```python
# Convert the data into a DataFrame
df = pd.DataFrame(data)
df
```
